```python
#importing the dependancies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
#loading the dataset to a pandas dataframe
creditcard_dataset= pd.read_csv('/content/creditcard.csv')
```

```python
#handling the missing values
creditcard_dataset.replace({'Amount':{'UNKNOWN':1}},inplace=True)
creditcard_dataset.replace({'Class':{'UNKNOWN':1}},inplace=True)
creditcard_dataset.replace({'V23,V24,V25,V26,V27,V28':{'UNKNOWN':0.1}},inplace=True)
```

```python
#sperating data and labels
x= creditcard_dataset[creditcard_dataset.Class == 0]
y= creditcard_dataset[creditcard_dataset.Class == 1]
```

```python
#balancing the data for accurate prediction
x_sample= x.sample(n=103)
```

```python
#preparing new dataset
new_dataset = pd.concat([x.sample(n=103), y], axis=0)
```

```python
#seperating data and labels
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the dataset into training and testing data
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model= LogisticRegression()
```

```python
#training the logistic regression model using training data
model.fit(X_train, Y_train)
```

```python
X_train_prediction= model.predict(X_train)
training_data_accuracy= f1_score(X_train_prediction, Y_train)
```

```python
print('training data accuracy:', training_data_accuracy)
```

```python
X_test_prediction= model.predict(X_test)
test_data_accuracy= f1_score(X_test_prediction, Y_test)
```

```python
print('test data accuracy:', test_data_accuracy)
```

```python
#improving accuracy using oversampling
import imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=2)
X_train_smote, Y_train_smote = smote.fit_resample(X_train, Y_train)
```

```python
log_reg_smote = LogisticRegression()
log_reg_smote.fit(X_train_smote, Y_train_smote)
```

```python
Y_pred_smote = log_reg_smote.predict(X_test)
accuracy_smote = f1_score(Y_test, Y_pred_smote)
print(f'Oversampling (SMOTE) Accuracy: {accuracy_smote}')
```

```python
#improving accuracy using undersampling
from imblearn.under_sampling import RandomUnderSampler
```

```python
undersampler = RandomUnderSampler(random_state=2)
X_train_under, Y_train_under = undersampler.fit_resample(X_train, Y_train)
```

```python
log_reg_under = LogisticRegression()
log_reg_under.fit(X_train_under, Y_train_under)
```

```python
Y_pred_under = log_reg_under.predict(X_test)
accuracy_under = f1_score(Y_test, Y_pred_under)
print(f'Undersampling Accuracy: {accuracy_under}')
```
